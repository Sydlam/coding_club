### A How To on Fork Syncing and Pushing to Your Branch

## Updating Your Exercises Folder

### Step 1

```python
git remote add upstream https://github.com/dunsmoorlab/coding_club.git
```

This line of code simply links the Dunsmoor lab coding club folder as your 'parent' or upstream branch

### Step 2

```python
git merge upstream/master
```

The command git merge upstream/master is used to merge changes from the master branch of a remote repository named upstream into your current local branch. This is commonly done when you are working on a fork of a repository and want to incorporate changes made in the original repository's master branch into your own branch.

Here's a step-by-step explanation of what this command does:

git merge: This is the Git command used to merge changes from one branch into another.

upstream: This is the name of the remote repository from which you want to pull changes. In Git, you can have multiple remotes configured for a repository, and upstream is just a common name used to refer to the original or upstream repository from which you forked your own repository.

master: This is the name of the branch in the remote repository (upstream in this case) from which you want to merge changes. It's assumed that the remote repository has a branch named master.

When you run this command, Git will attempt to merge the changes from upstream/master into your current local branch. If there are any conflicts between your local changes and the changes from upstream/master, Git will prompt you to resolve those conflicts manually.

## Pushing Your Changes to the Remote Fork

### Step 1 : Add

```python
git add .
```

Adds all the changed files into a "package" think of this process as filling a box with your changes and sending it through the mail.

### Step 2 : Commit

```python
git commit -m "Your Message You'd Like to Attach"
```

Adds a message to the package of files that can describe the changes made in brief. 

### Step 3 : Push

```python
git push
```

Sends the package of changes to your remote forked branch. 

### Step 4 :  Status

```python
git status
```
