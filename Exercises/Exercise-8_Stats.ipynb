# <span style="color:darkorange"><b><center>Statistical Analysis<center>

| Analysis Approach        | R Code                          | Explanation                                       |
|-------------------------|--------------------------------|---------------------------------------------------|
| Descriptive Statistics   | `summary(data)`                | Provides summary statistics (mean, median, etc.) |
|                         | `mean(data)`                   | Calculates the mean of a numeric variable.        |
|                         | `median(data)`                 | Calculates the median of a numeric variable.      |
|                         | `sd(data)`                     | Computes the standard deviation.                 |
|                         | `var(data)`                    | Computes the variance.                           |
|                         | `quantile(data, probs)`        | Calculates specified quantiles.                  |
|                         | `hist(data)`                   | Creates a histogram of the data.                 |
| Correlation Analysis    | `cor(data1, data2)`             | Calculates the correlation between two variables. |
|                         | `cor.test(data1, data2)`        | Performs a hypothesis test for correlation.       |
| T-Test                  | `t.test(data, mu = expected)`   | Tests if the sample mean is equal to `expected`.  |
| ANOVA                   | `aov(response ~ factor, data)`  | Analyzes variance between groups.                |
| Chi-Square Test         | `chisq.test(table)`             | Tests the association between categorical data.   |
| Linear Regression       | `lm(y ~ x, data)`               | Fits a linear regression model.                   |
| Logistic Regression     | `glm(y ~ x, family = binomial)` | Fits a logistic regression model.                |
| Nonparametric Tests     | `wilcox.test(data1, data2)`     | Performs a Wilcoxon rank-sum test.               |
|                         | `kruskal.test(response ~ factor, data)` | Performs Kruskal-Wallis test.           |
