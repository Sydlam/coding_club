## Relevant formulas

- square area: $s = (2 r)^2$
- circle area: $c = \pi r^2$
- $c/s = (\pi r^2) / (4 r^2) = \pi / 4$
- $\pi = 4 * c/s$

## Image to visualize the concept

![Darts](https://raw.githubusercontent.com/coderefinery/jupyter/main/example/darts.svg)

#### The Monte Carlo method is based on the principle that the ratio of the areas of two geometric shapes is proportional to the ratio of the points randomly distributed within those shapes. In the case of estimating π, you're comparing the ratio of the area of the quarter-circle to the area of the unit square. Since the quarter-circle's area is π/4 and the unit square's area is 1, the ratio of these areas is π/4.

#### By simulating random points within the unit square and checking how many fall inside the quarter-circle, you're approximating the ratio of their areas. Multiplying the resulting ratio by 4 gives you an estimate of π, as mentioned in the previous explanation.



```python
# importing modules that we will need

import random
import matplotlib.pyplot as plt
```

```python
# initializing the number of "throws"

num_points = 1000
```

```python
# here we "throw darts" and count the number of hits

points = []
hits = 0
for _ in range(num_points):
    x, y = random.random(), random.random()
    if x*x + y*y < 1.0:
        hits += 1
        points.append((x, y, "red"))
    else:
        points.append((x, y, "blue"))
```

```python
# unzip points into 3 lists
x, y, colors = zip(*points)

# define figure dimensions
fig, ax = plt.subplots()
fig.set_size_inches(6.0, 6.0)
plt.title('Darts Thrown') #add title

# plot results
ax.scatter(x, y, c=colors)
```

```python
# compute and print the estimate

fraction = hits / num_points
4 * fraction
```
